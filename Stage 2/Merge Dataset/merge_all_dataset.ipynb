```python
def check_duplicate_columns(df):
    """
    Memeriksa kolom duplikat dalam DataFrame pandas.

    Args:
        df (pd.DataFrame): DataFrame yang ingin diperiksa.

    Returns:
        dict: Dictionary yang berisi nama kolom duplikat sebagai kunci
              dan list nama kolom lain yang identik dengannya sebagai nilai.
              Mengembalikan dictionary kosong jika tidak ada kolom duplikat.
    """
    duplicate_columns = {}
    seen_columns = {}

    for col in df.columns:
        current_column = df[col].values.tobytes()  # Representasikan kolom sebagai bytes untuk perbandingan

        if current_column in seen_columns:
            if current_column not in duplicate_columns:
                duplicate_columns[current_column] = [seen_columns[current_column]]
            duplicate_columns[current_column].append(col)
        else:
            seen_columns[current_column] = col

    # Ubah kunci dictionary dari representasi bytes menjadi nama kolom pertama yang ditemukan
    readable_duplicates = {}
    for byte_repr, cols in duplicate_columns.items():
        first_col_name = [name for byte, name in seen_columns.items() if byte == byte_repr][0]
        readable_duplicates[first_col_name] = cols

    return readable_duplicates
```

```python
duplicates = check_duplicate_columns(merge_data)
```

```python
def drop_duplicate_columns(df, duplicates_dict):
    """
    Menghapus kolom duplikat dari DataFrame berdasarkan dictionary duplikat.

    Args:
        df (pd.DataFrame): DataFrame untuk menghapus kolom.
        duplicates_dict (dict): Dictionary kolom duplikat yang dihasilkan
                                 oleh fungsi check_duplicate_columns.

    Returns:
        pd.DataFrame: DataFrame dengan kolom duplikat dihapus (mempertahankan kolom pertama dari setiap set duplikat).
    """
    columns_to_drop = set()
    for first_col, duplicate_list in duplicates_dict.items():
        for dup_col in duplicate_list:
            columns_to_drop.add(dup_col)

    df_cleaned = df.drop(columns=list(columns_to_drop), axis=1)
    return df_cleaned

# Menggunakan fungsi untuk menghapus duplikat:
if duplicates:
    df_cleaned = drop_duplicate_columns(merge_data, duplicates)
    print("\nDataFrame setelah menghapus kolom duplikat:")
    df_cleaned
else:
    print("\nTidak ada kolom duplikat untuk dihapus.")
```

# Missing Values

```python
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)

    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : 'Percent Missing (%)'})

    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    'Percent Missing (%)', ascending=False).round(2)

    # Print some summary information
    print ('Your selected dataframe has ' + str(df.shape[1]) + ' columns.\n'
            'There are ' + str(mis_val_table_ren_columns.shape[0]) + ' columns that have missing values.')

    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

```python
missing = missing_values_table(df_cleaned)
missing
```

```python
# Hapus kolom dengan missing values di atas 20%
threshold = 20  
high_missing_cols = missing[missing['Percent Missing (%)'] > threshold].index.tolist()
cleaned = df_cleaned.drop(columns=high_missing_cols)
missing_values_table(cleaned)
```
